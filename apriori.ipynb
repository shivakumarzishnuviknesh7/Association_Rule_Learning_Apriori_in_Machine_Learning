```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)# no column name we have indicated using header = none
# apyori will expect certain form in the data so data frame doesn't have we have to recreate
# dataframe -> list
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])# i is row ,j  is column , all the values should be string in the list then only we can train and make it to learn
```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
# min_support -  3 times transaction a day in a week  so 3*7/7501 = 0.0027 -> 0.003
# min_confidence - 0.2 which is 20%
# min_lift - minimum 3 will give relevalt result not less than 3
# min_length - minimum number of element we want to have in our rule left to right, we set 2 , 1 in left 1 in right(buy 1 get 1)
# max_length - maximum number of element we want to have in our rule left to right we set 2 , 1 in left 1 in right
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
```

```
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
